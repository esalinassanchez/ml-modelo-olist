## El tiempo de entrega es muy importante, por lo que nos enfocamos en los registros donde order_delivered_customer_date, no sea nula ##

```python
df_delivery = df[df['order_status'] == 'delivered']
df_delivery
```

```python
df_delivery = df[df['order_delivered_customer_date'].notnull()].copy()
df_delivery
```

###Traduciendo los campos para los tiempos que pasan por un determinado estado:

- order_purchase_timestamp â†’ fecha_hora_compra
- order_approved_at â†’ fecha_aprobacion_pedido
- order_delivered_carrier_date â†’ fecha_entrega_transportista
- order_delivered_customer_date â†’ fecha_entrega_cliente
- order_estimated_delivery_date â†’ fecha_entrega_estimada
- shipping_limit_date â†’ fecha_limite_envio

###Analizando mÃ©tricas de tiempo de entrega

- Tiempo hasta aprobaciÃ³n (time_to_approve): order_approved_at - order_purchase_timestamp
- Tiempo hasta envÃ­o al transportista (time_to_ship): order_delivered_carrier_date - order_approved_at
- Tiempo hasta entrega al cliente (time_to_customer): order_delivered_customer_date - order_delivered_carrier_date
- Retraso respecto a la fecha estimada (delivery_delay): order_delivered_customer_date - order_estimated_delivery_date

```python
df_delivery['time_to_approve'] = (df_delivery['order_approved_at'] - df_delivery['order_purchase_timestamp']).dt.days
df_delivery['time_to_ship'] = (df_delivery['order_delivered_carrier_date'] - df_delivery['order_approved_at']).dt.days
df_delivery['time_to_customer'] = (df_delivery['order_delivered_customer_date'] - df_delivery['order_delivered_carrier_date']).dt.days
df_delivery['delivery_delay'] = (df_delivery['order_delivered_customer_date'] - df_delivery['order_estimated_delivery_date']).dt.days
```

```python
# Grafiquemos esto

time_cols = ['time_to_approve', 'time_to_ship', 'time_to_customer', 'delivery_delay']

plt.figure(figsize=(15,10))
for i, col in enumerate(time_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df_delivery[col].dropna(), bins=50, kde=True)
    plt.title(col)
    plt.xlabel("DÃ­as")
plt.tight_layout()
plt.show()
```

Hacer anÃ¡lisis por categorÃ­as relevantes, para encontrar dÃ³nde se concentran las demoras:

Por estado del cliente (customer_state)

Por estado del vendedor (seller_state)

Por categorÃ­a de producto (product_category_name_english)

```python
print(df_delivery.groupby('customer_state')['time_to_customer'].mean().sort_values(ascending=False))
print(df_delivery.groupby('seller_state')['time_to_customer'].mean().sort_values(ascending=False))
print(df_delivery.groupby('product_category_name_english')['time_to_customer'].mean().sort_values(ascending=False))
```

## Periodos de tiempo donde se puede apreciar mas demoras o menos ##

```python
df_delivery['order_purchase_timestamp'] = pd.to_datetime(df_delivery['order_purchase_timestamp'])
df_delivery['order_delivered_customer_date'] = pd.to_datetime(df_delivery['order_delivered_customer_date'])
```

```python
df_delivery['purchase_year'] = df_delivery['order_purchase_timestamp'].dt.year
df_delivery['purchase_month'] = df_delivery['order_purchase_timestamp'].dt.month
df_delivery['purchase_dayofweek'] = df_delivery['order_purchase_timestamp'].dt.dayofweek  # 0=Lunes, 6=Domingo
df_delivery['purchase_week'] = df_delivery['order_purchase_timestamp'].dt.isocalendar().week
```

```python
time_by_month = df_delivery.groupby('purchase_month')['time_to_customer'].mean()
time_by_month.plot(kind='bar', figsize=(10,5), color='cornflowerblue')
plt.title('Tiempo promedio de entrega al cliente por mes')
plt.ylabel('DÃ­as promedio')
plt.xlabel('Mes')
plt.xticks(rotation=0)
plt.show()
```

```python
delay_by_day = df_delivery.groupby('purchase_dayofweek')['delivery_delay'].mean()
delay_by_day.index = ['Lun', 'Mar', 'MiÃ©', 'Jue', 'Vie', 'SÃ¡b', 'Dom']  # para mejor visualizaciÃ³n

delay_by_day.plot(kind='bar', figsize=(8,4), color='salmon')
plt.title('Retraso promedio en la entrega segÃºn dÃ­a de la semana de compra')
plt.ylabel('DÃ­as promedio de retraso')
plt.xlabel('DÃ­a de la semana')
plt.show()
```

```python
orders_by_week = df_delivery.groupby('purchase_week').size()
orders_by_week.plot(figsize=(12,5))
plt.title('Cantidad de pedidos por semana del aÃ±o')
plt.xlabel('Semana del aÃ±o')
plt.ylabel('NÃºmero de pedidos')
plt.show()
```
