```python
# Calcular distancia entre pares de lat/lon (en km) usando haversine
import re
import numpy as np

def _normalize_col(name):
    s = name.lower()
    # quitar palabras lat/lon/latitude/longitude/lng/long
    s = re.sub(r'latitude|longitude|long|lng|lat|lon', '', s)
    s = re.sub(r'[^a-z0-9_]', '', s)
    s = s.strip('_')
    return s

# Detectar columnas候 que parezcan latitud y longitud
cols = list(df.columns)
lat_candidates = [c for c in cols if 'lat' in c.lower() or 'latitude' in c.lower()]
lon_candidates = [c for c in cols if any(k in c.lower() for k in ['lon','lng','long','longitude'])]

pairs = []
# Emparejar por normalización del nombre (prefijo común)
for lat in lat_candidates:
    for lon in lon_candidates:
        if _normalize_col(lat) == _normalize_col(lon):
            pairs.append((lat, lon))

# Si no se detectaron pares por nombre, intentar emparejar por posición si hay exactamente 2 candidatos en total
if not pairs and len(lat_candidates) == 1 and len(lon_candidates) == 1:
    pairs = [(lat_candidates[0], lon_candidates[0])]

# Si aún no hay pares detectados, intentar emparejar por prefijos separados por '_'
if not pairs and lat_candidates and lon_candidates:
    for lat in lat_candidates:
        lat_pref = lat.split('_')[:-1]
        for lon in lon_candidates:
            lon_pref = lon.split('_')[:-1]
            if lat_pref and lon_pref and lat_pref == lon_pref:
                pairs.append((lat, lon))

print('Columnas lat detectadas:', lat_candidates)
print('Columnas lon detectadas:', lon_candidates)
print('Pares detectados:', pairs)

def haversine_series(lat1, lon1, lat2, lon2):
    # espera series pandas o arrays; devuelve distancia en km
    R = 6371.0
    lat1_rad = np.radians(lat1.astype(float))
    lon1_rad = np.radians(lon1.astype(float))
    lat2_rad = np.radians(lat2.astype(float))
    lon2_rad = np.radians(lon2.astype(float))
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    a = np.sin(dlat/2.0)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

distance_cols = []
for lat_col, lon_col in pairs:
    # Intentar convertir a numérico y manejar nulos
    lat_series = pd.to_numeric(df[lat_col], errors='coerce')
    lon_series = pd.to_numeric(df[lon_col], errors='coerce')
    # Buscar un posible compañero (por ejemplo seller vs customer). Nombrado por la normalización
    base = _normalize_col(lat_col) or 'pair'
    dist_name = f'distance_{base}_km'
    df[dist_name] = haversine_series(lat_series, lon_series, lat_series, lon_series) if False else None
    # La línea anterior es un placeholder. Queremos la distancia entre dos puntos diferentes;
    # si los pares representan buyer/seller deben estar en columnas distintas.
    # Recalcular correctamente: asumimos que lat_col y lon_col son del mismo punto; por eso intentamos encontrar la pareja opuesta
    # Buscaremos por convención si el nombre contiene 'customer' y existe 'seller_lat'/'seller_lng' etc.
    # Intentar buscar pareja opuesta (buyer vs seller)
    alt_prefixes = ['seller', 'seller', 'seller', 'customer', 'buyer']
    # Buscar explícitamente columnas típicas
    candidate_pairs = []
    # Si normalización vacía, usar raw names
    # Construir búsqueda de pareja: reemplazar el fragmento lat/latitude por posible lon variants of other actor
    # En la mayoría de datasets hay columnas 'customer_lat','customer_lng' y 'seller_lat','seller_lng'
    # Si el par detectado viene de, por ejemplo, 'customer_lat'+'customer_lng', buscaremos 'seller_lat'+'seller_lng'
    norm = _normalize_col(lat_col)
    if 'customer' in lat_col.lower() or 'buyer' in lat_col.lower():
        # proponer seller como pareja
        for s in ['seller', 'seller_lat', 'seller_lng', 'seller_long', 'seller_longitude']:
            pass
    # Búsqueda práctica: si existen columnas 'seller_lat' y 'seller_lng' usarlas
    found = False
    possible_lat_names = [n for n in cols if re.search(r'seller.*lat|.*seller.*latitude', n.lower())] or [n for n in cols if re.search(r'seller', n.lower()) and 'lat' in n.lower()]
    possible_lon_names = [n for n in cols if re.search(r'seller.*(lon|lng|long|longitude)', n.lower())] or [n for n in cols if re.search(r'seller', n.lower()) and any(k in n.lower() for k in ['lon','lng','long'])]
    if possible_lat_names and possible_lon_names:
        # usar la primera pareja seller encontrada
        other_lat = possible_lat_names[0]
        other_lon = possible_lon_names[0]
        try:
            lat1 = pd.to_numeric(df[lat_col], errors='coerce')
            lon1 = pd.to_numeric(df[lon_col], errors='coerce')
            lat2 = pd.to_numeric(df[other_lat], errors='coerce')
            lon2 = pd.to_numeric(df[other_lon], errors='coerce')
            df[f'distance_{_normalize_col(lat_col)}_to_{_normalize_col(other_lat)}_km'] = haversine_series(lat1, lon1, lat2, lon2)
            distance_cols.append(f'distance_{_normalize_col(lat_col)}_to_{_normalize_col(other_lat)}_km')
            found = True
        except Exception as e:
            print('Error calculando distancia para', lat_col, other_lat, e)
    if not found:
        # Si no se encontró pareja opuesta, calcular distancia entre lat/lon de la misma fila no tiene sentido (0), así que omitimos
        print(f'No se encontró pareja opuesta para {lat_col}/{lon_col}; omitido')

print('Columnas de distancia creadas (si las hay):', distance_cols)
if distance_cols:
    display(df[distance_cols].head())
    print(df[distance_cols].describe())
```

```python
# Calcular volumen y densidad de los productos
# Columnas usadas: product_weight_g, product_length_cm, product_height_cm, product_width_cm
import numpy as np

w_col = 'product_weight_g'
l_col = 'product_length_cm'
h_col = 'product_height_cm'
wd_col = 'product_width_cm'

missing = [c for c in [w_col,l_col,h_col,wd_col] if c not in df.columns]
if missing:
    print('Faltan columnas para volumen/densidad:', missing)
else:
    # Convertir a numérico (coerce para convertir valores extraños a NaN)
    w = pd.to_numeric(df[w_col], errors='coerce')
    l = pd.to_numeric(df[l_col], errors='coerce')
    h = pd.to_numeric(df[h_col], errors='coerce')
    wd = pd.to_numeric(df[wd_col], errors='coerce')

    # Considerar dimensiones válidas solo si son > 0
    valid_dims = (l > 0) & (h > 0) & (wd > 0)

    # Volumen en cm^3 y en litros
    df['product_volume_cm3'] = np.where(valid_dims, (l * h * wd), np.nan)
    df['product_volume_l'] = df['product_volume_cm3'] / 1000.0

    # Densidad: g/cm^3 (y kg/m^3)
    df['product_density_g_cm3'] = np.where(df['product_volume_cm3'].notna() & w.notna() & (df['product_volume_cm3']>0), w / df['product_volume_cm3'], np.nan)
    df['product_density_kg_m3'] = df['product_density_g_cm3'] * 1000.0

    created = ['product_volume_cm3','product_volume_l','product_density_g_cm3','product_density_kg_m3']
    print('Columnas creadas:', created)
    display(df[created].head())
    print('Estadísticas de volumen y densidad:')
    print(df[created].describe())
```

```python
# Selección automática de features (pipeline):
# Filtrado por nulos/varianza, correlación, VIF (si disponible), MI y RandomForest+permutation importance
import numpy as np
from sklearn.feature_selection import mutual_info_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.inspection import permutation_importance
import warnings
warnings.filterwarnings('ignore')

target = 'delivery_time_days'
if target not in df.columns:
    raise ValueError(f'Target {target} no existe en el dataframe. Ejecuta la celda que crea el target primero.')

# Excluir columnas irrelevantes/IDs/fechas/target
exclude_patterns = ['_id', 'id', 'order_id', 'customer_id', 'product_id', 'seller_id']
exclude = [c for c in df.columns if any(p in c.lower() for p in exclude_patterns)]
exclude += [target, 'delivery_delay_days', 'order_purchase_timestamp', 'order_delivered_customer_date', 'order_estimated_delivery_date']
candidates = [c for c in df.columns if c not in exclude]
print('Candidates iniciales:', len(candidates))

# 1) Filtrar por nulos: eliminar columnas con >80% nulos
missing_pct = df[candidates].isnull().mean()
candidates = [c for c in candidates if missing_pct.get(c,0) <= 0.8]
print('Tras filtrar por nulos (>80%):', len(candidates))

# 2) Separar num/cat candidatos y quitar constantes/low variance
num_cols = [c for c in candidates if pd.api.types.is_numeric_dtype(df[c])]
cat_cols = [c for c in candidates if c not in num_cols]
# quitar num con varianza cero
num_cols = [c for c in num_cols if df[c].nunique() > 1]
print('Numéricas candidatas:', len(num_cols), 'Categóricas candidatas:', len(cat_cols))

# 3) Correlación: eliminar una de cada par con corr abs > 0.95 (numéricas)
to_drop_corr = set()
if len(num_cols) > 1:
    corr = df[num_cols].corr().abs()
    upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    for col in upper.columns:
        high = [c for c in upper.index if upper.loc[c,col] > 0.95]
        if high:
            # drop the column with higher mean correlation (heurística)
            col_mean = upper[col].mean()
            for h in high:
                h_mean = upper[h].mean()
                if col_mean >= h_mean:
                    to_drop_corr.add(col)
                else:
                    to_drop_corr.add(h)

num_cols = [c for c in num_cols if c not in to_drop_corr]
print('Tras filtrar por correlación (numéricas):', len(num_cols))

# 4) VIF (si está disponible)
try:
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    vif_drop = True
    X_vif = df[num_cols].copy()
    X_vif = X_vif.fillna(X_vif.median())
    vif = pd.Series([variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])], index=X_vif.columns)
    high_vif = vif[vif > 10].sort_values(ascending=False)
    print('VIF alto (si hay):', high_vif.head())
    # quitar columnas con VIF muy alto (iterativo si se quiere).
    num_cols = [c for c in num_cols if c not in high_vif.index]
except Exception as e:
    print('Statsmodels no disponible o error calculando VIF, se omite VIF:', e)

print('Numéricas tras VIF (o sin VIF):', len(num_cols))

# 5) Preparar X para MI y model importance: one-hot para categóricas de baja cardinalidad
cat_small = [c for c in cat_cols if df[c].nunique() <= 100]
cat_large = [c for c in cat_cols if c not in cat_small]
print('Categóricas small (one-hot):', len(cat_small), 'large (omitidas de momento):', len(cat_large))
feat_for_model = num_cols + cat_small
X = df[feat_for_model].copy()
X = pd.get_dummies(X, columns=cat_small, dummy_na=True) if cat_small else X
y = df[target]
mask = y.notna()
X = X[mask]
y = y[mask]
# Rellenar NA con mediana para este cálculo
X = X.fillna(X.median())

# Muestreo si es muy grande
sample_frac = 0.2 if len(X) > 50000 else 1.0
if sample_frac < 1.0:
    X_s = X.sample(frac=sample_frac, random_state=0)
    y_s = y.loc[X_s.index]
else:
    X_s = X
    y_s = y

# 6) Mutual information (num + encoded cats)
try:
    mi = mutual_info_regression(X_s, y_s, random_state=0)
    mi_ser = pd.Series(mi, index=X_s.columns).sort_values(ascending=False)
    print('Top 20 mutual info:')
    display(mi_ser.head(20))
except Exception as e:
    print('Error en mutual_info:', e)

# 7) RandomForest importance + permutation importance
rf_imp = None
perm_imp = None
try:
    X_train, X_val, y_train, y_val = train_test_split(X_s, y_s, test_size=0.2, random_state=0)
    rf = RandomForestRegressor(n_estimators=200, n_jobs=-1, random_state=0)
    rf.fit(X_train, y_train)
    rf_imp = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)
    print('Top 20 RF importances:')
    display(rf_imp.head(20))
    try:
        perm = permutation_importance(rf, X_val, y_val, n_repeats=5, random_state=0, n_jobs=-1)
        perm_imp = pd.Series(perm.importances_mean, index=X_val.columns).sort_values(ascending=False)
        print('Top 20 Permutation importances:')
        display(perm_imp.head(20))
    except Exception as e:
        print('No se pudo calcular permutation importance:', e)
except Exception as e:
    print('Error entrenando RandomForest (tal vez demasiadas features o datos):', e)

# 8) Combinar scores y generar shortlist
scores = pd.DataFrame(index=X_s.columns)
if 'mi_ser' in locals():
    scores['mi'] = mi_ser.reindex(scores.index).fillna(0)
if rf_imp is not None:
    scores['rf'] = rf_imp.reindex(scores.index).fillna(0)
if perm_imp is not None:
    scores['perm'] = perm_imp.reindex(scores.index).fillna(0)
if not scores.empty:
    # normalizar columnas y promediarlas
    scores = scores.apply(lambda s: (s - s.min()) / (s.max() - s.min()) if (s.max() - s.min())>0 else 0)
    scores['mean_score'] = scores.mean(axis=1)
    shortlist = scores.sort_values('mean_score', ascending=False).head(30).index.tolist()
else:
    shortlist = feat_for_model[:30] if feat_for_model else []

print('Shortlist (top features):', len(shortlist))
print(shortlist)

# Guardar shortlist
import csv
with open('feature_shortlist.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['feature'])
    for s in shortlist:
        writer.writerow([s])

print('Shortlist guardada en feature_shortlist.csv')
```

```python
# KPIs para tiempo de entrega: métricas y gráficas justificativas
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

# Ruta de salida para resumen de KPIs
out_csv = Path('kpi_delivery_summary.csv')

if 'delivery_time_days' not in df.columns:
    print('No existe delivery_time_days. Ejecuta la celda que crea el target primero.')
else:
    ser = df['delivery_time_days'].dropna()
    n = len(ser)
    mean = ser.mean()
    median = ser.median()
    p90 = np.percentile(ser, 90)
    p95 = np.percentile(ser, 95)
    std = ser.std()
    cv = std / mean if mean else np.nan
    pct_neg = (ser < 0).mean() * 100
    pct_long = (ser > p95).mean() * 100

    kpi_dict = {
        'count': n,
        'mean_days': mean,
        'median_days': median,
        'p90_days': p90,
        'p95_days': p95,
        'std_days': std,
        'cv': cv,
        'pct_negative_days': pct_neg,
        'pct_>p95_days': pct_long
    }
    # On-time metrics si existe delivery_delay_days
    if 'delivery_delay_days' in df.columns:
        dd = df['delivery_delay_days'].dropna()
        kpi_dict['avg_delay_days'] = dd.mean()
        kpi_dict['pct_on_time'] = (dd <= 0).mean() * 100
        kpi_dict['pct_late'] = (dd > 0).mean() * 100

    kpi_df = pd.DataFrame.from_dict(kpi_dict, orient='index', columns=['value'])
    kpi_df.to_csv(out_csv)
    print('KPIs calculados y guardados en', out_csv)
    display(kpi_df)

    # Grafica 1: histograma + líneas (mean/median/p90)
    try:
        plt.figure(figsize=(10,4))
        sns.histplot(ser, bins=80, kde=True, color='C0')
        plt.axvline(mean, color='k', linestyle='--', label=f'mean={mean:.2f}')
        plt.axvline(median, color='g', linestyle='-.', label=f'median={median:.2f}')
        plt.axvline(p90, color='r', linestyle=':', label=f'p90={p90:.2f}')
        plt.title('Histograma de delivery_time_days')
        plt.legend()
        plt.show()
    except Exception as e:
        print('Error graficando histograma:', e)

    # Grafica 2: CDF
    try:
        plt.figure(figsize=(8,4))
        sorted_ser = np.sort(ser.values)
        p = np.arange(1, len(sorted_ser)+1) / len(sorted_ser)
        plt.plot(sorted_ser, p, marker='.', linestyle='none')
        plt.xlabel('delivery_time_days')
        plt.ylabel('CDF')
        plt.title('Función de distribución acumulada (CDF)')
        plt.grid(True, alpha=0.3)
        plt.show()
    except Exception as e:
        print('Error graficando CDF:', e)

    # Grafica 3: Tendencia mensual (mediana) si hay timestamps
    try:
        ts_col = None
        for c in ['order_delivered_customer_date', 'order_purchase_timestamp', 'order_approved_at']:
            if c in df.columns:
                ts_col = c; break
        if ts_col:
            tmp = df[[ts_col, 'delivery_time_days']].dropna().copy()
            tmp[ts_col] = pd.to_datetime(tmp[ts_col], errors='coerce')
            tmp = tmp.dropna(subset=[ts_col])
            tmp['month'] = tmp[ts_col].dt.to_period('M')
            monthly = tmp.groupby('month')['delivery_time_days'].median().astype(float)
            plt.figure(figsize=(10,4))
            monthly.plot(marker='o')
            plt.title('Mediana de delivery_time_days por mes')
            plt.ylabel('median days')
            plt.xlabel('month')
            plt.xticks(rotation=45)
            plt.grid(alpha=0.3)
            plt.show()
    except Exception as e:
        print('Error calculando tendencia mensual:', e)

    # Grafica 4: Relación con distancia si existe columna distance
    try:
        dist_cols = [c for c in df.columns if (c.startswith('distance_') and c.endswith('_km')) or c=='distance']
        if dist_cols:
            d = dist_cols[0]
            mask = df[d].notna() & df['delivery_time_days'].notna()
            if mask.sum() > 0:
                sample = df.loc[mask, [d, 'delivery_time_days']].sample(n=min(5000, mask.sum()), random_state=0)
                plt.figure(figsize=(6,4))
                sns.scatterplot(x=d, y='delivery_time_days', data=sample, alpha=0.5, s=10)
                sns.regplot(x=d, y='delivery_time_days', data=sample, scatter=False, color='r')
                plt.title(f'Delivery time vs {d}')
                plt.show()
                print('Correlación (Pearson):', sample[d].corr(sample['delivery_time_days']))
        else:
            print('No se encontró columna de distancia para analizar relación con delivery_time')
    except Exception as e:
        print('Error analizando relación con distancia:', e)

    # Grafica 5: Boxplot por mes de compra (si existe columna month derivada)
    try:
        if 'order_purchase_timestamp_month' in df.columns:
            pmonth = 'order_purchase_timestamp_month'
            tmp = df[[pmonth, 'delivery_time_days']].dropna()
            # limitar a últimas 12 entradas únicas si existen muchas
            uniq = sorted(tmp[pmonth].unique())
            last = uniq[-12:] if len(uniq)>12 else uniq
            plt.figure(figsize=(12,5))
            sns.boxplot(x=pmonth, y='delivery_time_days', data=tmp[tmp[pmonth].isin(last)])
            plt.title('Boxplot delivery_time_days por mes de compra (últimos 12)')
            plt.xticks(rotation=45)
            plt.show()
    except Exception as e:
        print('Error boxplot por mes:', e)

    # Grafica 6: Mediana por categoría (si existe)
    try:
        if 'product_category_name' in df.columns:
            tmp = df[['product_category_name', 'delivery_time_days']].dropna()
            med = tmp.groupby('product_category_name')['delivery_time_days'].median().sort_values()
            top10 = med.tail(10).index.tolist()
            plt.figure(figsize=(8,4))
            med.loc[top10].plot(kind='barh')
            plt.title('Mediana delivery_time_days por categoría (top10 más lentas)')
            plt.xlabel('median days')
            plt.show()
    except Exception as e:
        print('Error analizando por categoría:', e)

    # Grafica 7: On-time rate mensual si existe delivery_delay_days
    try:
        if 'delivery_delay_days' in df.columns and ts_col:
            tmp = df[[ts_col, 'delivery_delay_days']].dropna().copy()
            tmp[ts_col] = pd.to_datetime(tmp[ts_col], errors='coerce')
            tmp = tmp.dropna(subset=[ts_col])
            tmp['month'] = tmp[ts_col].dt.to_period('M')
            on_time = tmp.groupby('month')['delivery_delay_days'].apply(lambda x: (x<=0).mean())
            plt.figure(figsize=(10,4))
            on_time.plot(marker='o')
            plt.title('On-time rate (mensual)')
            plt.ylabel('on-time fraction')
            plt.xlabel('month')
            plt.show()
    except Exception as e:
        print('Error calculando on-time rate mensual:', e)

    # Razonamiento impreso para justificar KPIs
    print('Razonamiento para selección de KPIs:')
    print('- mean/median/p90/p95: resumen del centro y extremos; p90/p95 ayudan a focalizar reducción de la cola larga')
    print('- pct_on_time / avg_delay: métricas operativas alineadas con SLA y satisfacción')
    print('- pct_negative_days: entregas tempranas que pueden indicar ruido o problemas en timestamps')
    print('- CV (coef. de variación): mide estabilidad del proceso; útil para priorizar reducción de varianza')
    print('- Tendencias mensuales: detectan estacionalidad o cambios en el proceso')
    print('- Relación con distancia y categoría: variables predictoras y accionables para operaciones')

    print('Propuesta KPI final para monitorización (sugerida):')
    print('1) Median delivery time (días) - robusto frente a outliers')
    print('2) 90th percentile delivery time (días) - objetivo para reducir cola larga')
    print('3) On-time rate (%) - medida de SLA/operaciones')
    print('4) Average delay (días) - magnitud de retrasos')
    print('5) % entregas tardías (delivery_delay_days > 0)')
    print('Se guardó', out_csv, 'con métricas básicas para reportes rápidos')
```
