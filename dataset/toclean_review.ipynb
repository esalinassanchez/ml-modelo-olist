```python
import pandas as pd

# Leer CSV completo
df = pd.read_csv(
    'olist_order_reviews_dataset.csv',
    sep=',',
    quotechar='"',
    encoding='utf-8',
    dtype=str,
    on_bad_lines='skip'  # Ignora filas que rompen el parseo
)

# Columnas de texto que pueden tener saltos de línea o comillas internas
text_cols = ['review_comment_title', 'review_comment_message']

for col in text_cols:
    # Reemplazar saltos de línea y retornos de carro por espacio
    df[col] = df[col].str.replace(r'[\r\n]+', ' ', regex=True)
    # Reemplazar comillas internas " por comillas dobles "" (escape de CSV)
    df[col] = df[col].str.replace('"', '""', regex=False)

# Guardar CSV limpio
df.to_csv(
    'olist_reviews_clean.csv',
    index=False,
    sep=',',
    quotechar='"',
    quoting=1  # Todos los campos de texto entre comillas
)
```
